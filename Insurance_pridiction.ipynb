# PROBLEM STATEMENT REFERRENCE.
Supervised machine learning is a type of machine learning where the model is trained on a labeled dataset. In this context, "labeled" means that each training example comes with a corresponding label or outcome that the model is supposed to predict. The goal is for the model to learn the relationship between the input data (features) and the output labels, so that it can make accurate predictions on new, unseen data.

### Key Concepts in Supervised Machine Learning:

1. **Labeled Data**:
   - **Features**: These are the input variables (independent variables) that are used to predict the output. In your case, features could include customer demographics, details of marketing calls, etc.
   - **Labels**: These are the output variables (dependent variables) that the model is trying to predict. For your problem, the label is whether a customer will subscribe to an insurance policy (binary: Yes/No or 1/0).

2. **Types of Supervised Learning**:
   - **Classification**: When the output variable is categorical (e.g., predicting whether a customer will subscribe or not), the problem is a classification problem. Examples of algorithms include Logistic Regression, Decision Trees, Random Forest, and Support Vector Machines (SVM).
   - **Regression**: When the output variable is continuous (e.g., predicting the exact amount of insurance premium), the problem is a regression problem. Examples of algorithms include Linear Regression, Ridge Regression, and Lasso Regression.

3. **Training**:
   - The model learns by comparing its predictions on the training data with the actual labels and adjusting its parameters to reduce the prediction error. This process is typically done using an optimization algorithm like Gradient Descent.

4. **Testing**:
   - After training, the model is evaluated on a separate set of data (testing data) that was not used during training. This helps to assess how well the model generalizes to new data.

5. **Model Evaluation Metrics**:
   - **Classification Metrics**: Accuracy, Precision, Recall, F1-Score, ROC-AUC.
   - **Regression Metrics**: Mean Absolute Error (MAE), Mean Squared Error (MSE), Root Mean Squared Error (RMSE), R-squared.

6. **Overfitting and Underfitting**:
   - **Overfitting**: When a model learns the training data too well, including noise and outliers, and performs poorly on new data.
   - **Underfitting**: When a model is too simple to capture the underlying patterns in the data, leading to poor performance on both training and testing data.

7. **Model Selection and Tuning**:
   - Choosing the right model and tuning its hyperparameters (settings that are not learned from the data) is crucial for optimizing performance. Techniques like cross-validation, grid search, and random search are commonly used for this purpose.

8. **Applications**:
   - Supervised learning is widely used in various industries for tasks like spam detection, sentiment analysis, fraud detection, image recognition, and, as in your case, predicting customer behavior.

### Example Workflow for a Supervised Learning Task:

1. **Data Collection**: Gather the data with input features and corresponding labels.
2. **Data Preprocessing**: Clean the data, handle missing values, encode categorical variables, and normalize or scale features.
3. **Feature Selection/Engineering**: Select or create the most relevant features for the task.
4. **Model Selection**: Choose a suitable model based on the nature of the problem (e.g., classification or regression).
5. **Training**: Train the model using the training data.
6. **Evaluation**: Evaluate the model on the test data using appropriate metrics.
7. **Hyperparameter Tuning**: Optimize the model's performance by tuning its hyperparameters.
8. **Deployment**: Deploy the model to make predictions on new data.
9. **Monitoring**: Monitor the model's performance over time and update it as necessary.

This process helps in building a predictive model that can assist in decision-making, such as determining the likelihood of a customer subscribing to an insurance policy based on their characteristics and previous interactions.

Remove Outliers using IQR

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
import numpy as np
Q1 = np.percentile(data['Age'], 25)
Q3 = np.percentile(data['Age'], 75)

# Calculate IQR
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
filtered_data = data['Age'][(data['Age'] >= lower_bound) & (data['Age'] <= upper_bound)]

```

```python
filtered_data
```

```python
sns.histplot(filtered_data, kde=True, bins=30, color='blue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot boxplot without outliers
plt.boxplot(filtered_data, vert=False)
plt.title('Box Plot of Age')
plt.ylabel('Age')
plt.show()
```

```python
# Drop rows where 'age' is an outlier
data.drop(data[(data['Age'] < lower_bound) | (data['Age'] > upper_bound)].index, inplace=True)
```

```python
cat_lis = ['Job', 'Marital', 'Education_Qualification']

# Loop through the categorical columns
for i in cat_lis:
    x = data[i]

    # Get unique labels in the column
    unique_labels = list(x.value_counts().index)

    # Create a dictionary for label encoding starting from 1
    label_mapping = {label: idx + 1 for idx, label in enumerate(unique_labels)}

    # Replace the original labels with the encoded ones
    data[i] = x.map(label_mapping)

```

```python
# Dictionary to map 'months' and 'Result' to numbers
month_map = {'jan': 1, 'feb': 2, 'mar': 3, 'apr': 4, 'may': 5, 'jun': 6, 'jul': 7, 'aug': 8, 'sep': 9, 'oct': 10, 'nov': 11, 'dec': 12}
Result_map = {'no': 0, 'yes': 1}

# Replace the 'months' and 'Result' names with numbers
data['Month'] = data['Month'].map(month_map)
data['Result'] = data['Result'].map(Result_map)
```

### Model Building and Fine turning

#### Use Random Forest Classifier

1. **Create the Random Forest model.**
2. **Use cross-validation to find the optimum number of trees (`n_estimators`).**
3. **Plot the results to visualize the optimum number of trees.**
4. **Choose the number of trees with the highest cross-validation score.**
5. **Train the model using the optimal number of trees.**
6. **Evaluate the model.**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score

# Step 1: Define the range of estimators to test
n_estimators_range = range(10, 210, 10)  # Testing n_estimators from 10 to 200
cv_scores = []  # To store cross-validation scores

# Step 2: Iterate over different numbers of trees (n_estimators)
for n_estimators in n_estimators_range:
    rfc = RandomForestClassifier(n_estimators=n_estimators, random_state=42)
    scores = cross_val_score(rfc, X_train, y_train, cv=5, scoring='accuracy')
    cv_scores.append(np.mean(scores))  # Store the mean cross-validation score

# Step 3: Plot the results to find the optimum number of trees
plt.plot(n_estimators_range, cv_scores, marker='o')
plt.xlabel('Number of Trees (n_estimators)')
plt.ylabel('Cross-Validated Accuracy')
plt.title('Optimum Number of Trees for Random Forest')
plt.show()

# Step 4: Choose the number of trees with the highest cross-validation score
optimal_n_estimators = n_estimators_range[np.argmax(cv_scores)]
print(f"Optimal number of trees: {optimal_n_estimators}")

# Step 5: Train the Random Forest classifier with the optimal number of trees
rfc_optimal = RandomForestClassifier(n_estimators=optimal_n_estimators, random_state=42)
rfc_optimal.fit(X_train, y_train)

# Step 6: Evaluate the model on the test set
y_pred = rfc_optimal.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Test set accuracy: {accuracy*100:.2f}")
```
